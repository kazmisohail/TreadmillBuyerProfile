```python
num_features = []
cat_features = []

# to separate numerical and categorical features for graphical analysis
for col in data.columns:
    if data[col].dtype == 'object':
        cat_features.append(col)
    else:
        num_features.append(col)
```

### Distribution Plots for Numerical Features

```python
for feature in num_features:
    sns.histplot(data[feature], color="skyblue")
    plt.title(f"Distribution of {feature}")
    plt.show()
```

### Count Plots for Numerical Features

```python
for feature in num_features:
    sns.countplot(data=data, x=feature, palette="pastel")
    plt.title(f"Count Plot of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

### Box Plots for Numerical Features

```python
for feature in num_features:
    sns.boxplot(data=data, y=feature, palette="coolwarm")
    plt.title(f"Box Plot of {feature}")
    plt.ylabel(feature)
    plt.show()
```

### Count Plots for Categorical Features

```python
for feature in cat_features:
    sns.countplot(data=data, x=feature, palette="pastel")
    plt.title(f"Count of {feature}")
    plt.show()
```

### Bivariate Analysis

```python
# Product vs Gender
sns.countplot(data=data, x="Product", hue="Gender", palette="pastel")
plt.title("Product vs Gender")
plt.show()
```

```python
# Product vs MaritalStatus
sns.countplot(data=data, x="Product", hue="Gender", palette="muted")
plt.title("Product vs MaritalStatus")
plt.show()
```

```python
# Product vs Age
sns.boxplot(data=data, x="Product", hue="Age", palette="coolwarm")
plt.title("Product vs Age")
plt.show()
```

## 6. Outlier Detection:
* Check for the outliers by using the IQR method.

```python
for feature in num_features:
    q1 = data[feature].quantile(0.25)
    q3 = data[feature].quantile(0.75)
    iqr = q3 - q1
    outliers = data[(data[feature] < (q1 - 1.5 * iqr)) | (data[feature] > (q3 + 1.5 * iqr))]
    print(f"\nOutliers for {feature}: {len(outliers)}")
```

## 7. Conditional Probabilities:

```python
product_counts = data['Product'].value_counts()
print("Percentage of Customers by Product:\n", (product_counts / len(data)) * 100)
```

### product - gender

```python
total_customers = len(data)
male_customers = len(data[data['Gender'] == 'Male'])
percent_male_customers = (male_customers / total_customers) * 100
print(f"Percentage of Male customers purchasing a treadmill: {percent_male_customers:.2f} %")
```

```python
female_kp781 = len(data[(data['Gender'] == 'Female') & (data['Product'] == 'KP781')])
percent_female_kp781 = (female_kp781 / total_customers) * 100
print(f"Percentage of Female customers purchasing KP781: {percent_female_kp781:.2f} %")
```

```python
female_given_kp281 = len(data[(data['Product'] == 'KP281') & (data['Gender'] == 'Female')]) / len(data[data['Product'] == 'KP281'])
print(f"Probability of a customer being Female given that Product is KP281: {female_given_kp281:.2f}")
```

### product - age

```python
age_20s_30s = len(data[(data['Age'] >= 20) & (data['Age'] < 40)])
percent_age_20s_30s = (age_20s_30s / total_customers) * 100
print(f"Percentage of customers aged between 20s and 30s: {percent_age_20s_30s:.2f} %")
```

### product - income

```python
low_income_threshold = data['Income'].quantile(0.25)
high_income_threshold = data['Income'].quantile(0.75)
```

```python
low_income_customers = len(data[data['Income'] < low_income_threshold])
percent_low_income_customers = (low_income_customers / total_customers) * 100
print(f"Percentage of low-income customers purchasing a treadmill: {percent_low_income_customers:.2f} %")
```

```python
high_income_kp781 = len(data[(data['Income'] > high_income_threshold) & (data['Product'] == 'KP781')])
percent_high_income_kp781 = (high_income_kp781 / total_customers) * 100
print(f"Percentage of high-income customers purchasing KP781: {percent_high_income_kp781:.2f} %")
```

```python
high_income_given_kp781 = (high_income_kp781 / len(data[data['Product'] == 'KP781'])) * 100
print(f"Probability of high-income salary given Product is KP781: {high_income_given_kp781:.2f} %")
```

### product - fitness

```python
fitness_5_customers = len(data[data['Fitness'] == 5])
percent_fitness_5 = (fitness_5_customers / total_customers) * 100
print(f"Percentage of customers with Fitness Level 5: {percent_fitness_5:.2f} %")
```

```python
fitness_5_kp781 = len(data[(data['Fitness'] == 5) & (data['Product'] == 'KP781')])
percent_fitness_5_kp781 = (fitness_5_kp781 / total_customers) * 100
print(f"Percentage of customers with Fitness Level 5 purchasing KP781: {percent_fitness_5_kp781:.2f} %")
```

```python
fitness_5_given_kp781 = fitness_5_kp781 / len(data[data['Product'] == 'KP781'])
print(f"Probability of Fitness Level 5 given KP781: {fitness_5_given_kp781:.2f}")
```

### product - marital status

```python
partnered_customers = len(data[data['MaritalStatus'] == 'Partnered'])
percent_partnered_customers = (partnered_customers / total_customers) * 100
print(f"Percentage of partnered customers using treadmills: {percent_partnered_customers:.2f} %")
```
